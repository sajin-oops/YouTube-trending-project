```python
import os
import json
import psycopg2
from dotenv import load_dotenv
```

```python

load_dotenv()

def get_connection():
    return psycopg2.connect(os.getenv("DATABASE_URL"))






def create_raw_table(conn):
    with conn.cursor() as cur:
        cur.execute("""
            CREATE SCHEMA IF NOT EXISTS raw;

            CREATE TABLE IF NOT EXISTS raw.youtube_trending (
                video_id        TEXT,
                title           TEXT,
                channel_id      TEXT,
                channel_title   TEXT,
                category_id     TEXT,
                published_at    TIMESTAMP,
                description     TEXT,
                view_count      BIGINT,
                like_count      BIGINT,
                comment_count   BIGINT,
                duration        TEXT,
                fetched_at      TIMESTAMP,
                region_code     TEXT
            );
        """)
        conn.commit()
    print("✅ Raw table ready")

def load_videos(conn, videos):
    with conn.cursor() as cur:
        for v in videos:
            cur.execute("""
                INSERT INTO raw.youtube_trending VALUES (
                    %(video_id)s, %(title)s, %(channel_id)s,
                    %(channel_title)s, %(category_id)s, %(published_at)s,
                    %(description)s, %(view_count)s, %(like_count)s,
                    %(comment_count)s, %(duration)s, %(fetched_at)s,
                    %(region_code)s
                )
            """, v)
        conn.commit()
    print(f"✅ Loaded {len(videos)} rows into raw.youtube_trending")

if __name__ == "__main__":
    with open("raw_videos.json") as f:
        videos = json.load(f)

    conn = get_connection()
    create_raw_table(conn)
    load_videos(conn, videos)
    conn.close()
```
